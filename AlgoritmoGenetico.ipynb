# Objetivo: Maximizar o número de 1's em uma string binária
---



```python
# Importações
import random
import numpy as np
```

# Definindo um cromossomo

```python
def gera_cromossomo(tamanho_cromossomo):
    """Função para gerar um cromossomo binário aleatório
    Args:
        tamanho_cromossomo (int): Tamanho do cromossomo
    Returns:
        cromossomo (list): lista Cromossomo binário aleatório
    """
    cromossomo = []
    for i in range(tamanho_cromossomo):
        cromossomo.append(random.randint(0,1))
    return cromossomo
```

# Definindo a população

```python
def gera_populacao(tamanho_populacao, tamanho_cromossomo):
    """Função para gerar uma população inicial
    Args:
        tamanho_populacao (int): Tamanho da população
        tamanho_cromossomo (int): Tamanho do cromossomo
    Returns:
        populacao (list): lista de cromossomos binários
    """
    populacao = []
    for i in range(tamanho_populacao):
        populacao.append(gera_cromossomo(tamanho_cromossomo))
    return populacao
```

# Avaliação de Fitness

```python
def conta_1s(cromossomo):
    """Função para contar o número de 1s em um cromossomo binário
    Args:
        cromossomo (list): lista de cromossomo binários

    Returns:
        int: número de 1s no cromossomo
    """
    total = sum(cromossomo)
    return total
```

# Modelar a função de Seleção

```python
def selecao(populacao):
    """Funcao para selecionar o melhor cromossomo
    Args:
        populacao (list): lista de cromossomos
    Returns:
        individuo (list): lista binaria
    """
    # 1. Calcula a soma total do fitness
    soma_fitness = sum(conta_1s(ind) for ind in populacao)

    # 2. Sorteia um número entre 0 e a soma total
    limite = random.uniform(0, soma_fitness)

    # 3. Percorre a população acumulando o fitness
    acumulado = 0
    for individuo in populacao:
        acumulado += conta_1s(individuo)
        if acumulado >= limite:
            return individuo
```

# Realizando o Cruzamento

```python
def cruzamento(pai, mae):
    """ Funcao de cruzamento crossover
    Args:
        pai (list): lista binária de um cromossomo
        mae (list): lista binária de outro cromossomo
    Returns:
        filho1 (list): filho1
        filho2 (list): filho2
    """
    if len(pai) != len(mae):
        raise ValueError("Os pais devem ter o mesmo tamanho")

    # Seleciona um ponto de corte aleatório (exceto nas pontas)
    ponto = random.randint(1, len(pai) - 1)

    # Filho 1: parte do pai1 + parte do pai2
    filho1 = pai[:ponto] + mae[ponto:]
    # Filho 2: parte do pai2 + parte do pai1
    filho2 = mae[:ponto] + pai[ponto:]

    return filho1, filho2
```

# Realizando Mutação

```python
def mutacao (cromossomo, taxa_mutacao):
    """ Funcao de mutacao
    Args:
        cromossomo (list): lista binária de um cromossomo
        taxa_mutacao (float): taxa de mutacao
    Returns:
        cromossomo (list): lista binária de um cromossomo mutado
    """
    cromossomo_mutado = []
    for bit in cromossomo:
        if random.random() < taxa_mutacao:
            # Inverte: 0 vira 1, 1 vira 0
            cromossomo_mutado.append(1 - bit)
        else:
            cromossomo_mutado.append(bit)

    return cromossomo_mutado
```

# Algoritmo Genético

```python
def algoritmo_genetico_steady_state():
    # Parâmetros
    tamanho_populacao = 10
    tamanho_cromossomo = 8
    taxa_mutacao = 0.05
    geracoes = 50 # Steady-state costuma precisar de mais iterações, mas são mais rápidas

    populacao = gera_populacao(tamanho_populacao, tamanho_cromossomo)

    for geracao in range(geracoes):
        # 1. Ordenar a população pelo fitness (Melhores primeiro)
        populacao = sorted(populacao, key=conta_1s, reverse=True)

        melhor_da_geracao = populacao[0]
        if geracao % 5 == 0: # Printar de 5 em 5 para não poluir o console
            print(f"Geração {geracao}: Melhor Fitness = {conta_1s(melhor_da_geracao)} -> {melhor_da_geracao}")

        if conta_1s(melhor_da_geracao) == tamanho_cromossomo:
            print(f"Objetivo alcançado na geração {geracao}!")
            break

        # 2. Selecionar dois pais (usando a função de seleção existente)
        p1 = selecao(populacao)
        p2 = selecao(populacao)

        # 3. Cruzamento e Mutação
        f1, f2 = cruzamento(p1, p2)
        f1 = mutacao(f1, taxa_mutacao)
        f2 = mutacao(f2, taxa_mutacao)

        # 4. SUBSTITUIÇÃO STEADY-STATE:
        # Substituímos os dois PIORES da população (os últimos da lista ordenada)
        populacao[-2] = f1
        populacao[-1] = f2

    # Resultado Final
    populacao = sorted(populacao, key=conta_1s, reverse=True)
    melhor_final = populacao[0]
    print(f"\nMelhor cromossomo final: {melhor_final} com {conta_1s(melhor_final)} 1's")

algoritmo_genetico_steady_state()
```
